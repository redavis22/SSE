```python
data = pd.read_csv("nba_data.csv")
```

```python
#Predict Offensive rating based on Defensive stats: Defensive Rebounds, Steals, and Blocks
#Logically the more defensively focused performance should lead to a higher Defensive rating
```

```python

#print(data.head())#grabs the first 5 elements of the dataset
data = data [["DRB%","STL%","BLK%","Offense"]]   
print(data.head())# grabs the modified data
```

```python
predict = "Offense" 
X=np.array(data.drop([predict],1))
Y=np.array(data[predict])
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X,Y,test_size=0.1)

```

```python
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test)
print(acc)
```

```python
print('Co: \n', linear.coef_)#m in y=mx+b
print('Intercept: \n',  linear.intercept_)#y intercept b
```

```python
predictions = linear.predict(x_test)
for x in range (len(predictions)):
        print(predictions[x], x_test[x],y_test[x])
```
